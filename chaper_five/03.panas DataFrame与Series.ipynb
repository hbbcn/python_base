```python
import pandas as pd

# 创建DataFrame 方式1: 字典列表
df = pd.DataFrame([
    {"姓名":"王林","语文":80,"数学":90,"英语":70},
    {"姓名":"李慕婉","语文":60,"数学":80,"英语":93},
    {"姓名":"十三","语文":80,"数学":90,"英语":70},
])

# 方式2:：字典
df2 = pd.DataFrame({
    "姓名":["王林","李慕婉","十三"],
    "语文":[80,60,80],
    "数学":[90,80,90],
    "英语":[70,93,70]
})

# 方式3:列表
df3 = pd.DataFrame([
    ["王林",80,90,70],
    ["李慕婉",60,80,93],
    ["十三",80,90,70]
],columns=["姓名","语文","数学","英语"])

df4 = pd.DataFrame([
    ["王林",80,90,70],
    ["李慕婉",60,80,93],
    ["十三",80,90,70]
],columns=["姓名","语文","数学","英语"],index=["a","b","c"])

# 元组方式
df5 = pd.DataFrame(
    [
        ("王林",80,90,70),
        ("李慕婉",60,80,93),
        ("十三",80,90,70)
    ],
    columns=["姓名","语文","数学","英语"],
    index=["a","b","c"]
)

# 2.DataFrame常见属性 常见属性：columns index,columns,values,size,dtypes,shape
# df4.index.tolist() # index: 索引
# df4.columns.tolist() # columns: 列
# df4.values.tolist() # values: 值
# df4.size # size: 元素个数
# df5.dtypes # dtypes: 数据类型
# df5.shape # shape:: 数据维度
print(df5.shape)

```
