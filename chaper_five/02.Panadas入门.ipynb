```python
import pandas as pd

df = pd.DataFrame({
    "姓名":["小王","小张","小李","小赵"],
    "语文":[80,60,90,70],
    "数学":[90,80,80,90],
    "英语":[71,95,100,8]
})
# 统计计算
f"语文最高分：{df['语文'].max()},最低分{df['语文'].min()},平均分：{df['语文'].mean():.2f}"
f"数学最高分：{df['数学'].max()},最低分{df['数学'].min()},平均分：{df['数学'].mean():.2f}"
f"英语最高分：{df['英语'].max()},最低分{df['英语'].min()},平均分：{df['英语'].mean():.2f}"
```
